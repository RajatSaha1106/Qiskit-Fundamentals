```python
import math,random

import matplotlib.pyplot as plt
```

Here $\beta$ signifies inverse temperature dependence, much like $\beta=\frac{1}{k_{B}T}$

```python
N=8
beta=2
sigma=math.sqrt(beta/N)
n_steps=6
xend=1.0 


for step in range(n_steps):
    upsilon=[0.0]
    x=[0.0]

    for k in range (N):
        upsilon.append(random.gauss(upsilon[-1],sigma))
    for k in range (1,N+1):
        x.append(upsilon[k]+(xend-upsilon[-1])*k/float(N))
    
    plt.plot(x,range(N+1))
    plt.title(r"Pulling Back Random Walk: $\beta={2}$")
    plt.xlabel("Position")
    plt.ylabel("Step")
```
